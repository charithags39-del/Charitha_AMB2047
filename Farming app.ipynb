```
crops=0
points=0
while(True):
    print("Welcome to farm game")
    print("Choose the options\n1.Plant the seeds\n2.Water the crops\n3.Harvest the crops\n4.Exit")
    choice=int(input("Enter your choice: "))

    if choice==1:
        crops=crops+1
        print("Seeds planted")
    elif choice==2:
        if crops>=1:
            print("Watered the crops")
        else:
            print("No crops to water")
    elif choice==3:
        if crops>=1:
            print("Harvested the crops")
            points=crops*10
        else:
            print("No crops to harvest")
    elif choice==4:
        print("Total points:",points,"\nThanks for playing")
        break
    else:
        print("Enter valid input")
```

```
dict_fish={"Rohu":2,"Catla":4,"Koi":3}
total_money=0
random_fish=(list(dict_fish.keys()))
print(random_fish)
while(True):
    print("Choose the option")
    print("1.Catching the fish\n2.Displaying inventory\n3.Sell the fish\n4.Exit")
    choice=int(input("Enter your choice: "))
    if choice==1:
        a=input("Enter the fish name: ")
        b=int(input("Enter the number of fish: "))
        print("Fish caught")
        dict_fish[a]=dict_fish.get(a,0)+b
    elif choice==2:
        print(dict_fish)
    elif choice==3:
        c=input("Enter the fish name you Want to sell: ")
        d=int(input("Enter the number of fish: "))
        if c=="Rohu" or c=="Catla" or c=="Koi":
            total_money=total_money+(100*d)
            dict_fish[c]=dict_fish.get(c,0)-d
    elif choice==4:
        print("Thankyou for shopping")
        print("Total money earned:",total_money)
        break
    else:
        print("Enter valid input")
```
