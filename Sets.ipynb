```python
# operation on set
a=set(["a","b","e","f"])
b=set(["a","c","d","e"])
print(a)
print(b)
print("a union b is ",a.union(b))
print("intersection of a, b",a.intersection(b))
print('a-b',a.difference(b))
print("difference a-b",a-b)
print("diference b-a ",b-a)
print("difference b-a",b.difference(a))
print("symetric difference a-b",a.symmetric_difference(b))
print("symetric difference b-a",b.symmetric_difference(a))


```

```python
S1=set([10,20,30,40,50])
S2=set([1,2,3,4,5])
print(S1)
print(S2)
print("S1 union S2",S1.union(S2))
print("S2 union S1", S2.union(S1))
print("intersection of S1 and S2",S1.intersection(S2))
print("difference S1-S2",S1.difference(S2))
print("difference S2-S1",S2.difference(S1))
print("symmetric difference S1-S2",S1.symmetric_difference(S2))
print("symmetric difference S2-S1",S2.symmetric_difference(S1))
print("is 20 present in S1", 20 in S1)
print(S1.update(S2))
print(S2.update(S1))
print("Sum of elements in S2", sum(S2))
print("Average of elements in S2", sum(S2)/len(S2))
print("Maximum element in S1",max(S1))
print("Minimum element in S1", min(S1))
print("Adding 6 to S2",S2.add(6))
print(S1)
```
