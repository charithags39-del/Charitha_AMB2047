```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('Salary_Data.csv')
df
```

```python
x=df.iloc[:,0].values
y=df.iloc[:,-1].values
print(x)
print(y)
```

```python
#spitting the data set set into traing and testing set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
print(df.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#calculate the mean value of input and output
x_mean=np.mean(x_train)
y_mean=np.mean(y_train)
print(x_mean)
print(y_mean)
```

```python
# calculate intercept and coefficient
n=len(x_train)
num,den=0,0

for i in range(n):
  num+=(x_train[i]-x_mean)*(y_train[i]-y_mean)
  den+=(x_train[i]-x_mean)**2
b1=num/den
b0=y_mean-(b1*x_mean)
print('b1=',b1)
print('y-intercept,b0=',b0)
```

```python
#predictingresult on tested data base
y_pred_code=b0+b1*x_test
print(y_pred_code)
```

```python
#mean square error
n1=len(y_test)
mse=0
for i in range(n1):
  mse+=(y_test[i]-y_pred_code[i])**2
mse=mse/n1
print(mse)
```

```python
#root mean square error
rmse=np.sqrt(mse)
print(rmse)
```
