```python
def expr(a,b):
    try:
        c=a/b
        print(c)
    except ZeroDivisionError:
        print("divided by zero")
    except TypeError:
        print("type error")
    except NameError:
        print("value is not declared")
    except:
        print("something went wrong")
    else:
        print("executed without error!!")

a=5
b=0
c=expr(a,b)
```
