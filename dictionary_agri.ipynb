```python
# A farmer maintains a dictionary containing the names of crops and their respective yields (in quintals per acre).

crop_yield = {
    "Rice": 120,
    "Wheat": 85,
    "Maize": 60
}
```

```python
#find the number of crops cultivated by the farmer

num_crops = len(crop_yield)
print("Number of Crops Cultivated:", num_crops)
```

```python
#display the yield of wheat

print("Wheat Yield:", crop_yield["Wheat"])
```

```python
# add a crop pulses with yield 45

crop_yield["Pulses"] = 45
print(crop_yield)
```

```python
#update the yield of rice to 130

crop_yield["Rice"] = 130
print(crop_yield)
```

```python
#display all crop names cultivated by farmer

crop_yields = crop_yield.keys()
print("Crops Cultivated:", list(crop_yields))
```

```python
#display all crop yields
#
yields = crop_yield.values()
print("Crop Yields:", list(yields))
```

```python
#check whether 'Sugarcane' is cultivated by the farmer

print("Sugarcane" in crop_yield)
```

```python
# How do you get the yield of maize from the dictionary?

print("Maize Yield:", crop_yield["Maize"])

print("Maize Yield:", crop_yield.get("Maize"))
```

```python
#calculate total yield of all crops

total_yield = sum(crop_yield.values())
print("Total Yield of all Crops:", total_yield)
```

```python
#display each crop and its yield using loop

for crop, yield_value in crop_yield.items():
    print(crop, ":", yield_value)
```

```python
# copy of a dictionary

crop_yield2 = crop_yield.copy()
print("Copy of crop_yield:", crop_yield2)
```

```python
# Creating a dictionary using dict()
crops = dict({
    "Rice": {"Season": "Kharif", "Yield": "4 tons/ha", "Price": 40},
    "Wheat": {"Season": "Rabi", "Yield": "3 tons/ha", "Price": 35},
    "Maize": {"Season": "Kharif", "Yield": "5 tons/ha", "Price": 30}
})

# keys()
print("Crop Names:", crops.keys())
print(list(crops.keys()))

# values()
print("\n Crop Details:", crops.values())
print(list(crops.values()))

# items()
print("\nAll Crop Items:", crops.items())
print(list(crops.items()))

# get()
print("Rice Details:", crops.get("Rice"))

# setdefault()
crops.setdefault("Barley", {"Season": "Rabi", "Yield": "2 tons/ha", "Price": 28})
print("After setdefault:", crops)

# update()
crops.update({"Sugarcane": {"Season": "Annual", "Yield": "70 tons/ha", "Price": 5}})
print("After update:", crops)

# pop()
removed_crop = crops.pop("Maize")
print("Removed Crop:", removed_crop)

# popitem()
last_item = crops.popitem()
print("Last Removed Item:", last_item)

# copy()
crops_copy = crops.copy()
print("Copied Dictionary:", crops_copy)

# clear()
crops.clear()
print("After clear:", crops)
```
