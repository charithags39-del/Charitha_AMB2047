```
from typing import Counter
inventory={}
total_money=0

import random

fish_prices={
    "Rohu":12,
    "Catla":18,
    "Sinchu":25
}
print("\n--Fishing Game Menu--")
print("1. Catch Fish")
print("2. View Inventory")
print("3. Sell Fish")
print("4. Exit")

while True:
    choice=int(input("\nEnter your choice: "))
    if choice==1:
        fish=random.choice(list(fish_prices.keys()))
        inventory[fish]=inventory.get(fish,0)+1
        print(f"\nYou caught a {fish}!")

    elif choice==2:
        if inventory=={}:
            print("No fish !")
        else:
            print("\n--Inventory--")
            for fish,count in inventory.items():
                print(f"{fish}: {count}")
    elif choice==3:
        if inventory=={}:
            print("No fish !")
        else:
            earnings=0
            for fish,count in inventory.items():
                earnings+=fish_prices[fish]*count
            total_money+=earnings
            inventory.clear()
            print(f"You sold fish for ${earnings}")
    elif choice==4:
        print(f"Final money eaned: ${total_money}")
        print("Thank you for playing !")
        break
    else:
        print("Invalid choice !")

```
