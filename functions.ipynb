```python
#Scope of variables

def func(x):
  print("local x is:",x)
  x=2
  print("Changed local to:",x)

x=50
func(x)
print("x is still :",x)
```

```python
#local variable in function
def func():
  x=2
  print("Changed local to:",x)

x=50
func()
print("x is still :",x)
```

```python
def func1():
  global x
  print("x is ",x)
  x=2
  print("Changed local to:",x)

x=50
func1()
print("x is still :",x)
```

```python
#Required argument

def printme(xy):
  print(xy)
  #return

#printme() #statement gives out error
printme("Hello")
printme(1234)
```

```python
#Keyword argument

def printme(xy):
  print(xy)
  return


printme("Hello")
printme(xy="My string")
#printme(x="My string") #ERROR
```

```python
#keyword argument


def func(a,b=5,c=10):
  print("a=",a,"b=",b,"c=",c)

func(3,7)  # by default value will print acc to order
func(25,c=24)
func(c=50,a=100) # parameter can be of any order
```

```python
#default argument

def printinfo(name,age:35):
  print("Name:",name)
  print("Age=",age)

printinfo(age=50,name="Virat")
printinfo("Virat",23)
#printinfo(age=35) #ERROR
```
