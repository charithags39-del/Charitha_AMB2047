```python
pH=float(input("Enter the pH of soil"))
if(pH< 5.5):
  print("Acidic soil")
elif(pH>=5.5 and pH<=7.5):
  print("Neutral soil")
elif(pH>7.5 and pH<=8.5):
  print("Alkaline soil")
else:
  print("Highly alkaline soil")


```
