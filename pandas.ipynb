```python
import pandas as pd
production=pd.Series (
    [30,60,45,25],
          index=["wheat","rice","maize","barley"]
)
print(production)
```

```python
#by crop name (label)
print(production["rice"])
print(production.loc["rice"])    #loc=loaction by label
```

```python
#by positioning
print(production.iloc[0]) #iloc=indexing by location
```

```python
print(production.sum())
print(production.max())
print(production.mean())
print(production.idxmax()) #maximum production crop, it prints the index label of the maximum value in the Series
print(production.idxmax(),"is having max value of",production.max())
```

```python
# increase all crop production by10%(good harvest year)
increased_production=production*1.10
print(increased_production)
```

```python
#crop production greater than 40 tons
high_production=production[production>40]
print(high_production)
print()
print(production>40)
```

```python
import pandas as pd
data={
    "farm":["farmA","farmB","farmC","farmD"],
    "crop":["wheat","rice","wheat","maize"],
    "area_hectares":[10,15,8,12],
    "production_tons":[30,60,24,36]
}
df=pd.DataFrame(data)
print(df)
```

```python
df.head(2) #for first few rows
```

```python
df.tail(3) #for bottom few rows
```

```python
df.info()   #basic information about data type
```

```python
df.describe()   #summary of statestics
```

```python
#add a new coloumn (yield
#yield=production/area
df["yield_tons_per_hectare"]=df["production_tons"]/df["area_hectares"]
print(df)
```

```python
print(df.iloc[[1,3]])
```

```python
print(df.loc[[1,3],["farm","crop","area_hectares"]])
```

```python
df.iloc[[1,3],[0,1,2]]
```

```python
#Filter data
#farms growing wheat

wheat_farms=df[df["crop"]=="wheat"]
print(wheat_farms)
```

```python
#farms with yield>3

high_yield=df[df["yield_tons_per_hectare"]>3]
print(high_yield)
```

```python
#sort farms by highest production

sorted_df=df.sort_values(by="production_tons",ascending=False)
print(sorted_df)
```

```python
#grouped poeration-average yield per crop
avg_yield=df.groupby("crop")["yield_tons_per_hectare"].mean()
print(avg_yield)
```

```python
avg_yield=df.groupby("crop")["yield_tons_per_hectare"].sum()
print(avg_yield)
```
