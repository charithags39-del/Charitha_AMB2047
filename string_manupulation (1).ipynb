```python
#Extracting ALL strings with a prefix

codes=[
    "CRP_RICE_2026",
    "CRP_WHEAT_2025",
    "FRT_MANGO_2026",
    "CRP_MAIZE_2024",
]
filtered=[]

for crop_code in codes:
  if crop_code.startswith("CRP_"):
    filtered.append(crop_code)
print(filtered)
```

```python
#extract and clean in one go
codes=[
    "CRP_RICE_2026",
    "CRP_WHEAT_2025",
    "FRT_MANGO_2026",
    "CRP_MAIZE_2024",
]
clean_codes=[]

for code in codes:
  if code.startswith("CRP_"):
      clean_codes.append(code[4:])
print(clean_codes)
```

```python
#split method
codes=[
    "CRP_RICE_2026",
    "CRP_WHEAT_2025",
    "FRT_MANGO_2026",
    "CRP_MAIZE_2024",
]

for crop_code in codes:
  if crop_code.startswith("CRP_"):
    code,crop,year=crop_code.split("_")
    print(crop,year)
```

```python
#mini agriculture data cleaningexample
#a consolidated example using multiplestring operations.
raw_record="rice,1200,black soil**"
clean_record=raw_record.strip()
clean_record1=clean_record.rstrip("*")
crop,yield_kg,soil=clean_record1.split(",")
print(f"crop:{crop}")
print(f"yield:{yield_kg}kg")
print(f"soiltype:{soil}")
print(clean_record1)
```
