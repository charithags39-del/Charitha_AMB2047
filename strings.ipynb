```
codes=[
    "CRP_RICE_2026",
    "CRP_WHEAT_2025",
    "FRT_MANGO_2026",
    "CRP_MAIZE_2024"
]
filtered=[]
for code in codes:
  if code.startswith("CRP"):#True
    filtered.append(code)
print(filtered)
```

```
codes=[
    "CRP_RICE_2026",
    "CRP_WHEAT_2025",
    "FRT_MANGO_2026",
    "CRP_MAIZE_2024"
]
clean_codes=[]
for code in codes:
  if code.startswith("CRP_"):
    clean_codes.append(code[4:])
print(clean_codes)
```

```
codes=[
    "CRP_RICE_2026",
    "CRP_WHEAT_2025",
    "FRT_MANGO_2026",
    "CRP_MAIZE_2024"
]
for code in codes:
  if code.startswith("CRP_"):
    code,crop,year=code.split("_")
    print(crop,year)
```
